# **Introduction**

**Churn prediction** is one of the most widely used Big Data applications in the business world. It focuses on identifying customers who are likely to stop using or cancel a subscription-based service.

Although it was initially popular among large telecommunications companies, this approach is now relevant for businesses of all sizes, including startups. With the availability of predictive services and APIs, advanced analytics is no longer limited to large organizations with dedicated data science teams.

As an example of how churn prediction can support business growth, consider companies that operate on a subscription model, such as telecom providers, SaaS businesses, or any organization offering services with recurring monthly fees.

In general, there are three key strategies to increase revenue: acquiring new customers, increasing the value of existing customers through upselling, and improving customer retention. Each of these strategies involves certain costs, so the main focus should be on return on investment (ROI), which measures the balance between the additional revenue generated and the expenses incurred.

## Splitting Data

To prepare the data for model development, a portion of the dataset is selected as training data to be used as input for the machine learning algorithm. This process ensures that the trained classification model can generalize effectively to unseen data. In this study, 30% of the data is allocated as the sample size, which is considered an appropriate proportion for splitting between training and testing sets.

```python
from sklearn.model_selection import train_test_split
Y = train['Churn']
X = train.drop(columns=['Churn'])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=9)
```

```python
print('X train shape: ', X_train.shape)
print('Y train shape: ', Y_train.shape)
print('X test shape: ', X_test.shape)
print('Y test shape: ', Y_test.shape)
```

### **Random Forest Classification**

Random Forest is a supervised learning algorithm that builds an ensemble of decision trees. Typically, these trees are trained using the bagging (bootstrap aggregating) method. The core idea behind bagging is to combine multiple models that may individually be noisy but are approximately unbiased, in order to reduce overall variance and improve prediction performance.

The algorithm works as follows:

1. Let N represent the number of data instances and M the total number of features in the dataset.

2. At each node, a subset of features (m, where m < M) is randomly selected to determine the best split.

3. A bootstrap sample is drawn from the training data to build each individual tree, while the remaining data can be used to estimate error.

4. For each node in the tree, the algorithm selects the best split based only on the randomly chosen subset of features (m).

For prediction, a new data instance is passed through each tree in the forest until it reaches a terminal node, where it is assigned a class label. This process is repeated across all trees, and the final prediction is determined by majority voting among the trees.

In this study, the number of trees in the forest is set to 100.

```python
from sklearn.ensemble import RandomForestClassifier

# We define the model
rfcla = RandomForestClassifier(n_estimators=100,random_state=9,n_jobs=-1)

# We train model
rfcla.fit(X_train, Y_train)

# We predict target values
Y_predict5 = rfcla.predict(X_test)
```

```python
# The confusion matrix
rfcla_cm = confusion_matrix(Y_test, Y_predict5)
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(rfcla_cm, annot=True, linewidth=0.7, linecolor='black', fmt='g', ax=ax, cmap="BuPu")
plt.title('Random Forest Classification Confusion Matrix')
plt.xlabel('Y predict')
plt.ylabel('Y test')
plt.show()
```

```python
# Test score
score_rfcla = rfcla.score(X_test, Y_test)
print(score_rfcla)
```

## ROC Curve

The ROC Curve (Receiver Operating Characteristic Curve) is a graphical representation used to evaluate the performance of a binary classification model. It illustrates how well the classifier distinguishes between classes by plotting the trade-off between true positive rate and false positive rate at various threshold settings.

```python
from sklearn.metrics import roc_curve
# Random Forest Classification
Y_predict5_proba = rfcla.predict_proba(X_test)
Y_predict5_proba = Y_predict5_proba[:, 1]
fpr, tpr, thresholds = roc_curve(Y_test, Y_predict5_proba)
plt.subplot(331)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr, label='ANN')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('ROC Curve Random Forest')
plt.grid(True)
plt.subplots_adjust(top=2, bottom=0.08, left=0.10, right=1.4, hspace=0.45, wspace=0.45)
plt.show()

# Naive Bayes Classification
Y_predict3_proba = nbcla.predict_proba(X_test)
Y_predict3_proba = Y_predict3_proba[:, 1]
fpr, tpr, thresholds = roc_curve(Y_test, Y_predict3_proba)
plt.subplot(332)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr, label='ANN')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('ROC Curve Naive Bayes')
plt.grid(True)
plt.subplots_adjust(top=2, bottom=0.08, left=0.10, right=1.4, hspace=0.45, wspace=0.45)
plt.show()
```
